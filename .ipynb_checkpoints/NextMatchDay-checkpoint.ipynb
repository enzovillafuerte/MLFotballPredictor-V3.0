```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
pd.set_option('display.max_columns', None)
from datetime import datetime

def scrape_fbref_xG(url):
    
    # Creatind a dictionary of index and labels
    
    dict_index = { 
        2: '_Standard',
        4: '_GK',
        6: '_AdvGK',
        8: '_Shooting',
        10: '_Passing',
        12: '_PassTypes',
        14: 'G&SCreation',
        16: '_DefActions',
        18: '_Possession',
        20: '_PlayTime',
        22: '_Misc'
    }
    
    html_page = requests.get(url).text
    data = BeautifulSoup(html_page, 'html.parser')
    
    # Initializing an empty dataframe
    merged_df = None
    
    for index, label in dict_index.items():
        
        # Extract the specific table (assuming it's the 3rd table on the page)
        tables = data.find_all('table')
        rows = tables[index].find_all('tr')

        # Extract the headers
        headers = rows[1].find_all('th')
        column_titles = [header.get_text(strip=True) for header in headers]

        # Creating title label for accompanying the default naming convention
        ## IF statements here
        title_label = label

        # Add '_Standard' to each column name
        column_titles = [column_titles[0]] + [title + title_label for title in column_titles[1:]]

        # Check the number of columns expected
        print(f"Expected number of columns: {len(column_titles)}")

        # Extract the data rows
        table_data = []
        for row in rows[1:]:  # Start from the second row to skip the header row
            cols_a = [col.get_text(strip=True) for col in row.find_all('a')]
            cols_b = [col.get_text(strip=True) for col in row.find_all('td')]
            combined_cols = cols_a + cols_b

            # Print the length of combined columns for debugging
            # print(f"Row has {len(combined_cols)} columns: {combined_cols}")

            # Append only if the number of columns matches the headers
            # The first row was empty, that's why the error in the previous block of code. Keep this one
            if len(combined_cols) == len(column_titles):
                table_data.append(combined_cols)
            else:
                print(f"Skipping row with {len(combined_cols)} columns, expected {len(column_titles)}.")
    
        # Create DataFrame and set column titles
        df = pd.DataFrame(table_data, columns=column_titles)
        
        # Merge dataframe with the existing dataframe
        if merged_df is None:
            merged_df = df
        else:
            merged_df = pd.merge(merged_df, df, how='outer', on='Squad')
            
    # Adding an extraction date column
    today = datetime.today()
    merged_df['Extraction Date'] = today
    
    return merged_df

# Example usage
url = 'https://fbref.com/en/comps/9/Premier-League-Stats'
epl_df = scrape_fbref_xG(url)

epl_df
```

### Transformations

```python
excluded_cols = [col for col in merged_df.columns if '90' in col]
merged_filtered = merged_df.drop(columns=excluded_cols)
```

##### Defining which columns we should not apply transformations to

```python
## Columns that should not be divided by 90: Maybe need to add more or less columns?
no_divide_cols = ['# Pl_Standard_x', 'Age_Standard_x', 'Poss_Standard_x', 'MP_Standard_x', 'Starts_Standard_x', 
                  'Squad_y', '# Pl_Standard_y', 'Age_Standard_y', 'Poss_Standard_y', 'MP_Standard_y', 'Starts_Standard_y']

# Columns we don't want include when converting into floats
exclude_columns = ['Home', 'Away', 'Date', 'Match', 'Squad_x', 'Extraction Date_x', 'Squad_y', 'Extraction Date_y']
```

##### Converting objects to float

```python
#Some columns have - and + in the rows so to take that into account, remove them string characters
def clean_and_convert(value):
    if isinstance(value, str):
        value = value.strip()
        if value.startswith('+'):
            return float(value[1:])
        elif value.startswith('-'):
            return -float(value[1:])
    return pd.to_numeric(value, errors='coerce')


# Should we convert the Date into Date dtype or is it fine as a string? 
```

```python
# Apply clean_and_convert only to columns that are not in exclude_columns
for col in merged_filtered.columns:
    if col not in exclude_columns:
        merged_filtered[col] = merged_filtered[col].apply(clean_and_convert)
    else:
        # Ensure excluded columns remain as strings
        merged_filtered[col] = merged_filtered[col].astype(str)
```

##### Dividing all float columns by 90

```python
#Divide all other columns by 90
for col in merged_filtered.columns:
    if col not in no_divide_cols and col not in exclude_columns:
        merged_filtered[col] = merged_filtered[col] / 90
```
